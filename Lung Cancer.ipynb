```python
# To read Dataset with using Pandas.
import pandas as pa 

columns = [
    'First','Secound','Third','4th',
    '5th','6th','7th', '8th',
    '9th', '10th','11th', '12th',
    '13th', '14th', '15th', '16th', 
    '17th', '18th', '19th','20th',
    '21th', '22th','23th', '24th', 
    '25th', '26th', '27th', '29th',
    '30th', '31th', '32','33th',
    '34th', '35th', '36th', '37th',
    '38th', '39th', '40th','41th',
    '42th', '43th','44th', '45th',
    '46th', '47th', '48th', '49th',
    '50th', '51th','52th', '53th',
    '54th', '55th', '56th', '57th', 
    '58th'
]

data = pa.read_csv('lung-cancer.data', names = columns)
data.head()
```

```python
# To know types of dataset and know position of Missing values.
data.dtypes
```

```python
# Function to return number only.
def is_numeric(x):
    return x.isnumeric()
```

```python
# Mask to return number only.
mask = data['5th'].apply(is_numeric)

mask.head()
```

```python
# Remove some of Missing values from dataset.
numeric_data = data[mask]

numeric_data.head()
```

```python
print(len(data))
print(len(numeric_data))
```

```python
# Show the data applying the mask.
mask = data['40th'].apply(is_numeric)

mask.head()
```

```python
# Remove the last part of Missing values from dataset.
isnumeric_data = numeric_data[mask]

isnumeric_data.head()
```

```python
# Know the different between the data with ( missing values and without it ).
print(len(data))
print(len(isnumeric_data))
```

```python
isnumeric_data.dtypes
```

```python
# Change the data_type of columns that have missing values. 
data_type_cha = {'5th':'int64','40th':'int64'}

isnumeric_data = isnumeric_data.astype(data_type_cha)
```

```python
isnumeric_data.dtypes
```

```python
isnumeric_data.head()
```

```python
# Seperate data to two parts input and output.
data_input = isnumeric_data.drop(columns = ['58th'])
data_output = isnumeric_data['58th']
```

```python
# Show data input only.
data_input.head()
```

```python
# Show data output only.
data_output.head()
```

```python
# Using unique() to know the values of output.
data_output.unique()
```

```python
# Change the data of output to two part 0 and 1.
data_output = data_output.replace({2:0,1:1})

data_output.unique()
```

```python
# Show output before replace to 0 and 1.
data_output.head()
```

```python
# Seperate input and output to tree parts ( Training , Valdation , Testing ).
from sklearn.model_selection import train_test_split

x , x_test , y , y_test = train_test_split(data_input,data_output, test_size = 0.25 , random_state = 2 )
x_train , x_val , y_train , y_val = train_test_split(x,y,test_size = 0.25 , random_state = 2 )
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
# Call the model that you will train it.(DecisionTreeClassifier).
from sklearn.tree import DecisionTreeClassifier
```

```python
# Give the model paremeters (max_depth).
model = DecisionTreeClassifier(max_depth = 3 , random_state = 2)
```

```python
# Training the model.
model.fit(x_train,y_train)
```

```python
# Predict the output to give him the input and save the value the predict it.
y_predict_train = model.predict(x_train)
y_predict_val = model.predict(x_val)
```

```python
# Call the accuracy_score to know accuracy of model.
from sklearn.metrics import accuracy_score
```

```python
# Use accuracy_score to know the defrent between ( the real_value and predict_value ).
print(accuracy_score(y_train,y_predict_train))
print(accuracy_score(y_val,y_predict_val))
```

```python
# Make for loop to change the deepth of DecisionTree and know what the best result. 
max_depth_value =[1,2,3,4,5,6,7,8]
acc_train_values = []
acc_val_values = []

for max_depth in max_depth_value:
    model = DecisionTreeClassifier(max_depth = max_depth , random_state = 10)
    model.fit(x_train,y_train)
    y_predict_train = model.predict(x_train)
    y_predict_val = model.predict(x_val)
    acc_train = accuracy_score(y_train,y_predict_train)
    acc_val = accuracy_score(y_val,y_predict_val)
    acc_train_values.append(acc_train)
    acc_val_values.append(acc_val)
```

```python
print(acc_train_values)
print(acc_val_values)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(max_depth_value,acc_train_values, label = 'Acc Train')
plt.plot(max_depth_value,acc_val_values, label = "Acc Val")
plt.legend()
plt.grid(axis = 'both')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
plt.title('Effect of max depth on accuracy')
plt.show()
```

```python
# Know the best model and train it. 
best_model = DecisionTreeClassifier(max_depth = 1 , random_state = 2)
best_model.fit(x_train,y_train)
```

```python
# Predict the test of dataset and caluculate accuracy.
y_predict_test = best_model.predict(x_test)
print(accuracy_score(y_test,y_predict_test))
```
